```python
import pandas as pd
from pandas import read_csv
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

ds = read_csv("../data/heart.csv")

ds.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn import svm

for col in ds.select_dtypes(include='object').columns:
    ds[col] = ds[col].astype('category').cat.codes


train_X , test_X, train_Y, test_Y= train_test_split(ds.drop(columns="HeartDisease"), ds["HeartDisease"], test_size=0.25, random_state=5)

train_X.head(5)
test_X.head(5)
```

DECISION TREE


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

dst = DecisionTreeClassifier()
dst.fit(train_X, train_Y)
y_pred = dst.predict(train_X)

accuracy_score(train_Y, y_pred)
```

KNN


```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(train_X, train_Y)
```
